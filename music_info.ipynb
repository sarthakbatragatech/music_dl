```python
import os
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import librosa
import librosa.display

import IPython.display as ipd
```

## Load Audio

```python
sound_path = 'data/sample/space_oddity.wav'
y, sr = librosa.load(sound_path)
print(y.shape, sr)
```

```python
ipd.Audio(sound_path)
```

## Display Waverform

```python
librosa.display.waveplot(y, sr=sr);
```

## Fast Fourier Transform

```python
n_fft = 2048
D = np.abs(librosa.stft(
    y[:n_fft], 
    n_fft=n_fft,
    hop_length=n_fft+1))
plt.plot(D)
```

```python
Y = librosa.stft(y)
Ydb = librosa.amplitude_to_db(abs(Y))
plt.figure(figsize=(12, 5))
librosa.display.specshow(Ydb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
librosa.display.specshow(Ydb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
hop_length = 512
D = np.abs(librosa.stft(y, n_fft=n_fft,  
                        hop_length=hop_length))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='linear');
plt.colorbar();
```

```python
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=sr, hop_length=hop_length, 
                         x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
```
